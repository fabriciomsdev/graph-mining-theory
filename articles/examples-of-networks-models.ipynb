# Example of network models

```python
import networkx as nx
```

## Random network

```python
aleatory_network = nx.erdos_renyi_graph(100, 0.1)

nx.write_gml(aleatory_network, "aleatory_network.gml")
nx.draw(aleatory_network)
```

## Small world networks

```python
small_world_network = nx.watts_strogatz_graph(200, 4, 0.1)

nx.write_gml(small_world_network, "small_world_network.gml")

nx.draw(small_world_network)
```

### Free scale networks

```python
free_scale_network = nx.barabasi_albert_graph(200, 2)

nx.write_gml(free_scale_network, "free_scale_network.gml")

nx.draw(free_scale_network)
```
